**Dataset Description:**

From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

X = energy_data[['T2']]
y = energy_data['T6']


linreg1 = LinearRegression()
linreg1.fit(X, y)

print(round(linreg1.score(X, y), 2))
```

- Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. 
- The target variable is “Appliances”.
Use a 70-30 train-test set split with a random state of 42 (for reproducibility).
- Run a multiple linear regression using the training set and evaluate your model on the test set.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Drop columns
df = energy_data.drop(['date', 'lights'], axis=1)

y = df['Appliances']
X = df.drop('Appliances', axis=1)

scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)

X_scaled.head()
```

```python
# Drop columns
df = energy_data.drop(['date', 'lights'], axis=1)

# Instantiate scaler and perform transform
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)

# Pick out the target and feature sets
y = df_scaled['Appliances']
X_scaled = df_scaled.drop('Appliances', axis=1)

X_scaled.head()
```

```python
# Perform a 70/30 split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Perform Linear Regression analysis
linreg2 = LinearRegression()
linreg2.fit(X_train, y_train)

y_pred = linreg2.predict(X_test)

# Model Evaluation with metrics

MAE = mean_absolute_error(y_test, y_pred)
print(round(MAE,2))

RMSE =np.sqrt(mean_squared_error(y_test, y_pred))
print(round(RMSE, 3))

R2 = r2_score(y_test, y_pred)
print(round(R2, 2))

RSS = np.sum(np.square(y_test - y_pred))
print(round(RSS, 2))
```

```python
# Linear Model Weights
weights = pd.Series(linreg2.coef_, X_scaled.columns).sort_values()
weights_df = pd.DataFrame(weights).reset_index()

weights_df
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.4)


ridge.fit(X_train, y_train)

y_ridge = ridge.predict(X_test)

RMSE_ridge =np.sqrt(mean_squared_error(y_test, y_ridge))
print(RMSE_ridge)
```

```python
# Lasso Regression
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.001)


lasso.fit(X_train, y_train)

lasso_weights = pd.Series(lasso.coef_, X_scaled.columns).sort_values()
lasso_weights_df = pd.DataFrame(lasso_weights).reset_index()

lasso_weights_df
```

```python
y_lasso = lasso.predict(X_test)

RMSE_lasso =np.sqrt(mean_squared_error(y_test, y_lasso))
print(round(RMSE_lasso, 3))
```
